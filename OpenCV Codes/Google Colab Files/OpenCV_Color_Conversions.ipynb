**OpenCV Color Conversions**

`cv2.cvtColor()` Converts an image from one color space to another. The function converts an input image from one color space to another.Note that the default color format in OpenCV is often referred to as RGB but it is actually BGR (the bytes are reversed).

**Function Syntax**

dst = cv2.cvtColor( src, code )

where,
1. dst : It is the output image of the same size and depth as src.
2. src : input image
3. code : color space conversion code

**Commonly used color space conversion codes include:**

*   By Default it gives BGR Image
*   cv2.COLOR_BGR2RGB: Converts BGR (Blue, Green, Red) to RGB (Red,Green, Blue).
*   cv2.COLOR_BGR2GRAY: Converts BGR to grayscale.
*   cv2.COLOR_RGB2BGR: Converts RGB to BGR.
*   cv2.COLOR_BGR2HSV : Converts BGR to HSV(Hue, Saturation, and Value)
*   cv2.COLOR_RGB2HSV : Converts RGB to HSV









```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('28.jpg')
plt.imshow(img)
```

```python
# Read an image
img = cv2.imread('28.jpg')

# Convert BGR to RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
```

```python
# Convert BGR to GRAY
img_rgb1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.imshow(img_rgb1)
```

```python
# Convert RGB to BGR
img_rgb2 = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2BGR)

plt.imshow(img_rgb2)
```

```python
# Convert BGR to HSV
img_rgb3 = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

plt.imshow(img_rgb3)
```

```python
# Convert RGB to HSV
img_rgb4 = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2HSV)

plt.imshow(img_rgb4)
```
